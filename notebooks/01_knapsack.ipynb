```python
# Knapsack Notebook (alias-based, backend-agnostic test)

from optilab.aliases import Model, GRB, quicksum
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------------------------------------
# Model
# -------------------------------------------------------------------------------
m = Model()
print("Backend:", m.backend_name)

# -------------------------------------------------------------------------------
# Problem data
# -------------------------------------------------------------------------------
# Reproducibility (important for solver comparisons)
rng = np.random.default_rng(42)

# Problem size (scale this later)
n = rng.integers(15, 40)

# Random values and weights
values = rng.integers(1, 20, size=n)
weights = rng.integers(1, 15, size=n)

# Capacity = fraction of total weight (standard knapsack trick)
capacity = int(0.25 * np.sum(weights))

print("n items:", n)
print(f"   values : {values}")
print(f"   weights: {weights}")
print("capacity:", capacity)

# -------------------------------------------------------------------------------
# Decision variables
# -------------------------------------------------------------------------------
x = [m.add_binary_var(name=f"x_{i}") for i in range(n)]

# -------------------------------------------------------------------------------
# Constraint
# -------------------------------------------------------------------------------
m.add_constraint(
    quicksum(weights[i] * x[i] for i in range(n)) <= capacity
)

# -------------------------------------------------------------------------------
# Objective
# -------------------------------------------------------------------------------
m.set_objective(
    quicksum(values[i] * x[i] for i in range(n)),
    GRB.MAXIMIZE
)

# -------------------------------------------------------------------------------
# Solve
# -------------------------------------------------------------------------------
m.solve()

# -------------------------------------------------------------------------------
# Extract solution
# -------------------------------------------------------------------------------
solution = np.array([v.x for v in x])

print(f"Solution:   {solution}")
print("Total value:", float(np.sum(values * solution)))
print("Total weight:", float(np.sum(weights * solution)))
```
